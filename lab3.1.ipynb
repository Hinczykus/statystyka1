<div class="alert-info">

    
<br>  
    <font size = "+1"><b>Rozkład prawdopodobieństwa</b></font>
<br>
    
<b>Teoretyczny rozkład</b> może być zdefiniowany przez:<br>
* dystrybuantę $F(x)$
* gęstość lub funkcję masy prawdopodobieństwa $f(x)$<br>
    
<b>Empiryczny rozkład</b> wyznaczamy w oparciu o dane (próbę losową):<br>
* dystrybuanta empiryczna $\tilde{F}(x)$
* histogram $h(x)$<br>
    
<br>  
    <font size = "+1"><b>Teoretyczne parametry rozkładów prawdopodobieństwa</b></font>
<br>
 
 1. Wartość średnia (oczekiwana) $\,\mu=EX$<br>
 2. Wariancja $\,\sigma^2=Var(X)=E(X-\mu)^2$<br>
 3. Odchylenie standardowe $\,\sigma=std(X)=\sqrt(Var(X))$<br>
 4. Mediana $med(X)$
 5. Dominanta $dom(X)$<br>
    
<br>  
    <font size = "+1"><b>Estymatory parametrów rozkładów prawdopodobieństwa</b></font>
<br>

Estymatory pozwalają na oszacowanie wartośći nieznanego parametru rozkładu na podstwie próby losowej $(X_1,...,X_n)$. Np.:
 1. Średnia próbkowa (arytmetyczna) $\,\hat{\mu}=\bar{X}=\bar{X}(X_1,...,X_n)$<br>
    $$\bar{X}=\frac{1}{n}\sum_{i=1}^n X_i$$
 2. Wariancja próbkowa $\,\hat{\sigma}^2=Var(X_1,...,X_n)$<br>
    * estymator obciążony $$S^{2}=\frac{1}{n}\sum_{i=1}^n (X_i-\bar{X})^2$$
    * estymator nieobciążony $$S^{*2}=\frac{1}{n-1}\sum_{i=1}^n (X_i-\bar{X})^2$$
 
<br>  
    <font size = "+1"><b>Obciążenie estymatora</b></font>
<br>

Obciążenie estymatora $\hat{\theta}_n=\hat{\theta}_n(X_1,...,X_n)$ parametru $\theta$ definiujemy
$$B(\hat{\theta}_n)= E(\hat{\theta}_n)-\theta.$$
  
<br>

</div>


<div class="alert-info">

    
<br>  
    <font size = "+1"><b>Mieszanina dwóch rozkładów normalnych</b></font>
<br>
 
Niech zmienne losowe X i Y są niezależne i mają rozkłady normalne o następujących parametrach:<br>
$X \sim N( \mu_X,\sigma_X^2 ),$<br>
$Y \sim N( \mu_Y,\sigma_Y^2 ).$<br>
    
Rozpatrzmy nową zmienną losową Z, która jest kombinacją liniową X i Y dla dowonych skalarów $a,b$. Wtedy<br>
$Z = a\cdot X+b\cdot Y \sim N( \mu_Z,\sigma_Z^2).$<br>
<br>
Wartosć średnia zmiennej losowej Z wynosi
$EZ = \mu_Z = a\cdot\mu_X+b\cdot\mu_Y,$<br>
a wariancja $Var(Z)=\sigma_Z^2=a^2\cdot\sigma_X^2+b^2\cdot\sigma_Y^2.$<br>
<br>

</div>

<br>  
    <font size = "+1"><b>ZAD 1. Rozkład teoretyczny</b></font>
<br>
 
<br>

Oblicz średnią $EZ$ i wariancję $Var(Z)$ ze wzorów teoretycznych. <br>   
Wyrysuj na jednym rysunku gęstości prawdopodobieństwa zm. los. $X$, $Y$ i $Z$: 

(a) $\color{red}{\textbf{(2 p.)}}$ 
$X \sim N( 0,1 )$,  $Y \sim N( 2,1 )$,  $Z = 3\cdot X + 3\cdot Y $<br><br>

(b) $\color{red}{\textbf{(2 p.)}}$ 
$X \sim N( 1,1 )$,  $Y \sim N( 3,1 )$,  $Z = -3\cdot X + 2\cdot Y $<br><br>

(c) $\color{red}{\textbf{(2 p.)}}$ 
$X \sim N( 0,1 )$,  $Y \sim N( 3,4 )$,  $Z = 2\cdot X + Y $<br><br>
    
</div>

```python
# **Funkcja rysująca funkcję gęstości rozkładu normalnego N(mean,var)**

from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

def rysuj_gestosc(m, v):
    # parametry:
    # m - średnia
    # v - wariancja
    # sqrt(v) - odchylenie stand.
    
    x = np.linspace(-5,15,100) # wartości na osi OX
    g = norm.pdf(x, loc=m, scale=np.sqrt(v))# wartości funkcji gęstości na osi OY
    plt.plot(x, g, label=f"m={m}, var={v}")
    plt.xlabel('x')
    plt.ylabel('f ( x )')
    plt.legend()
    plt.grid(True)
    plt.title(f'Funkcja gęstości N(m,var)')
    return 
```

```python
# (a)
plt.figure()
rysuj_gestosc(0, 1)  # X
rysuj_gestosc(2, 1)  # Y
rysuj_gestosc(6, 18) # Z
plt.title('Zad 1(a): X~N(0,1), Y~N(2,1), Z=3X+3Y')
plt.show()
```

```python
# (b)
plt.figure()
rysuj_gestosc(1, 1)  # X
rysuj_gestosc(3, 1)  # Y
rysuj_gestosc(3, 13) # Z
plt.title('Zad 1(b): X~N(1,1), Y~N(3,1), Z=-3X+2Y')
plt.show()
```

```python
# (c)
plt.figure()
rysuj_gestosc(0, 1)  # X
rysuj_gestosc(3, 4)  # Y
rysuj_gestosc(3, 8) # Z
plt.title('Zad 1(b): X~N(1,1), Y~N(3,1), Z=-3X+2Y')
plt.show()
```

<br>


    
* Wygeneruj po $n=100$ próbek dla zmiennych losowych $X$ oraz $Y$ o rozkładach normalnych (parametry poniżej). 
<br>Wskazówka: przeczytaj o funkcji [norm.rvs()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html)
    
* Następnie oblicz $n$ próbek $(Z_1,...,Z_n)$ dla zmiennej losowej $Z$ korzystając ze wzoru: 
$Z = a\cdot X+b\cdot Y.$ <br>Wskazówka: Każda indywidualna próbka $Z_i$ jest mieszaniną próbek $X_i$ i $Y_i$.

* Wyestymuj wartości parametrów rozkładu za pomocą estymatorów na podstawie próby losowej $(Z_1,...,Z_n)$:<br>
    1. średnia próbkowa (arytmetyczna) $\,\bar{Z}=\bar{Z}(Z_1,...,Z_n)$,<br>
    2. wariancja próbkowa $\,S^2=S^2(Z_1,...,Z_n)$.<br>

Porównaj wyniki z teoretycznymi wartościami parametrów.<br>

(a) $\color{red}{\textbf{(1 p.)}}$ 
$X \sim N( 0,1 )$,  $Y \sim N( 2,1 )$,  $Z = 3\cdot X + 3\cdot Y $<br><br>

(b) $\color{red}{\textbf{(1 p.)}}$ 
$X \sim N( 1,1 )$,  $Y \sim N( 3,1 )$,  $Z = -3\cdot X + 2\cdot Y $<br><br>

(c) $\color{red}{\textbf{(1 p.)}}$ 
$X \sim N( 0,1 )$,  $Y \sim N( 3,4 )$,  $Z = 2\cdot X + Y $<br><br>

    
</div>

```python
from scipy.stats import norm

# PARAMETRY (a)
n = 100
mx = 0
sx = 1
my = 2 
sy = 1
a = 3 
b = 3

# GENEROWANIE PRÓB LOSOWYCH dla X, Y
data_x = norm.rvs(loc=mx, scale=sx, size=n)
data_y = norm.rvs(loc=my, scale=sy, size=n)

# WYZNACZENIE PRÓBY LOSOWEJ Z 
data_z = a * data_x + b * data_y

# ESTYMOWANIE WARTOŚCI PARAMETRÓW ZM. LOS. Z
est_m = np.mean(data_z)
est_s2 = np.var(data_z, ddof=1)
# Obliczenia teoretyczne
teo_m = a * mx + b * my 
teo_s2 = (a**2) * (sx**2) + (b**2) * (sy**2)
print(est_m)
print(est_s2)
print(teo_m)
print(teo_s2)
```

    <font size = "+1"><b>ZAD 3. Estymacja wariancji i obciążenie estymatora</b></font>
<br>
<br> 

Niech zmienna losowa X ma rozkład normalny $X\sim N(\mu,\sigma^2)$ o parametrach $\mu=3$ i $\sigma^2=4$. 

Wygeneruj dla zmiennej losowej $X$ próby losowe o licznościach $n=10,50,100,500,1000,5000,10000,50000,100000$ za pomocą funkcji  [norm.rvs()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html).<br><br>
    
Zakładamy następującą sytuację: 
* Mamy zestaw danych w postaci próby losowej $(X_1,X_2,...,X_n)$ o liczności $n$. 
* Przy jej wykorzystaniu chcemy wyestymować nieznane nam parametry rozkładu: średnią $\mu$ i wariancję $\sigma^2$.
* Chcemy prześledzić wpływ liczności próby losowej na jakość estymacji.<br><br>


</div>

```python
from scipy.stats import norm
# STAŁE PARAMETRY
licznosc = [10,50,100,500,1000,5000,10000,50000,100000] # liczności prób losowch n
mu = 3 # wartość oczekiwana (średnia)
sigma = 2 # wariancja sigma^2 = 4
for n in licznosc:
    proba = norm.rvs(loc=mu, scale=sigma, size=n)
```


<div class="alert-success">
<br>
  

<b>ZAD 3.1. Nieobciążony estymator średniej $\color{red}{\textbf{(2 p.)}}$</b>

$\bar{X}=\frac{1}{n}\sum_{i=1}^n X_i$<br>
 
Oblicz wartości średnie za pomocą estymatora średniej $\bar{X}$ dla różnych liczności próby $n$. <br>Wyrysuj na wykresie zależność tych wartości od $n$.<br><br>


</div>

```python
# ZAD 3.1. Nieobciążony estymator średniej - obliczenia (uzupełnij)

from scipy.stats import norm
import numpy as np
estimated_mean = [] # średnia próbkowa 

for n in licznosc:
    proba= norm.rvs(loc=mu,scale=sigma,size=n)
    srednia_probkowa = np.mean(proba)
    estimated_mean.append(srednia_probkowa)
```

```python
# ZAD 3.1. Nieobciążony estymator średniej - rysunek

import numpy as np
import matplotlib.pyplot as plt
estimated_mean = [] # średnia próbkowa 

for n in licznosc:
    proba= norm.rvs(loc=mu,scale=sigma,size=n)
    srednia_probkowa = np.mean(proba)
    estimated_mean.append(srednia_probkowa)
theoretical_mean = [mu]*len(licznosc)
plt.plot(licznosc, estimated_mean, 'o-', label='est. $\hat{\mu}$')
plt.plot(licznosc, theoretical_mean, 'r-', label='teor. $\mu$')
plt.xscale('log') # oś OX logarytmiczna
plt.title('Estymator średniej $\hat{\mu}$')
plt.xlabel('Liczność próby n')
plt.ylabel('$\hat{\mu}$')
plt.grid(True)
plt.legend(loc="best", frameon=True)
plt.show()
```


<div class="alert-success">

<b>ZAD 3.2. Estymatory wariancji $\color{red}{\textbf{(2 p.)}}$</b>

* Obciążony estymator wariancji: $S^{2}=\frac{1}{n}\sum_{i=1}^n (X_i-\bar{X})^2$<br> 
* Nieobciążony estymator wariancji: $S^{*2}=\frac{1}{n-1}\sum_{i=1}^n (X_i-\bar{X})^2$<br>    
    
Oblicz wartości wariancji za pomocą estymatorów wariancji dla różnych liczności próby $n$. <br>Wyrysuj na jednym wykresie zależność tych wartości od $n$.<br><br>
    

</div>

```python
# ZAD 3.2. Nieobciążony i obciążony estymator wariancji - obliczenia (uzupełnij)

from scipy.stats import norm

estimated_var = []  # estymator wariancji S2
estimated_var_unbiased = [] # estymator wariancji S2*

for n in licznosc:
        # Obliczenie obciążonego estymatora wariancji (S²)
    proba= norm.rvs(loc=mu,scale=sigma,size=n)
    var_biased = np.var(proba, ddof=0)
    estimated_var.append(var_biased)
    
    # Obliczenie nieobciążonego estymatora wariancji (S²*)
    var_unbiased = np.var(proba, ddof=1)
    estimated_var_unbiased.append(var_unbiased)
```

```python
# ZAD 3.2. Nieobciążony i obciążony estymator wariancji - rysunek

import numpy as np
import matplotlib.pyplot as plt
estimated_var = []  # estymator wariancji S2
estimated_var_unbiased = [] # estymator wariancji S2*

for n in licznosc:
        # Obliczenie obciążonego estymatora wariancji (S²)
    proba= norm.rvs(loc=mu,scale=sigma,size=n)
    var_biased = np.var(proba, ddof=0)
    estimated_var.append(var_biased)
    
    # Obliczenie nieobciążonego estymatora wariancji (S²*)
    var_unbiased = np.var(proba, ddof=1)
    estimated_var_unbiased.append(var_unbiased)

theoretical_var = [sigma**2]*len(licznosc)
plt.plot(licznosc, estimated_var, 'o-', label='obciążony $S^2$')
plt.plot(licznosc, estimated_var_unbiased, 'o-', label='nieobciążony $S^{*2}$')
plt.plot(licznosc, theoretical_var, 'r-', label='$\sigma^2$')
plt.xscale('log') # oś OX logarytmiczna
plt.title('Estymatory wariancji')
plt.xlabel('Liczność próby n')
plt.grid(True)
plt.legend(loc='best', frameon=True)
plt.show()
```

<b>ZAD 3.3. Obciążenie estymatora wariancji</b>
    
$B(S^2)=S^2 - E(S^2)$<br>
    
Oczywiście tutaj $E(S^2)=\sigma^2=4$
    
* Oblicz obciążenie estymatora $B(S^2)$ dla różnych liczności próby $n$. 
* Czy asymptotyczna wartość obciążenia dąży do zera, tzn. $lim_{n\rightarrow + \infty} B(S^2) = 0$?<br><br>


</div>

```python
# ZAD 3.3. Obciążenie estymatora - obliczenia

from scipy.stats import norm

bias = [] # obciążenie B(S2) estymatora wariancji S2 
theoretical_bias = [] # obciążenie teor.

for n in licznosc:
    # generate random numbers from N(mu,sigma^2)
    data_normal = norm.rvs( size=n,loc=mu,scale=sigma )
    avg = sum( data_normal )/n
    expr = sum( (data_normal-avg)**2 ) 
    bias.append( abs( (expr/n)-sigma**2 ) ) 
    theoretical_bias.append( abs( -sigma**2/n) )
```

```python
# ZAD 3.3. Obciążenie estymatora - rysunek

import numpy as np
import matplotlib.pyplot as plt

plt.plot(licznosc, bias, 'o-', label='est. $B(S^2)$')
plt.plot(licznosc, theoretical_bias, 'r-', label='teor. $B(S^2)$')
plt.xscale('log') # oś OX logarytmiczna
plt.title('Obciążenie estymatora wariancji $S^2$')
plt.xlabel('Liczność próby n')
plt.ylabel(' $B ( S^2 )$ ')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
