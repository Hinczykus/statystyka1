<b>Rozkłady pradopodobieństwa</b>

Podział ze względu na wartości, jakie przyjmuje zmienna losowa:
* rozkłady ciągłe (np. rozkład cechy typu wzrost lub waga w populacji) -- [SciPy continuous distribution](https://docs.scipy.org/doc/scipy/tutorial/stats/continuous.html)
* rozkłady dyskretne (np. liczba oczek w rzucie kostką) -- [SciPy discrete distribution](https://docs.scipy.org/doc/scipy/tutorial/stats/discrete.html)
 
Podział ze względu na liczbę zmiennych losowych:
* rozkład jednowymiarowe (dla pojedynczej zmiennej losowej)
* rozkład wielowymiarowe (dla wektora losowego)

<br>

    
<b>Dystrubuanta</b> 

Dystrybuantą (ang. cumulative distribution function, CDF) jednowymiarowego rozkładu prawdopodobieństwa nazywamy funkcję $F: \mathbb{R} \to \mathbb{R}$ zdefiniowaną wzorem:
$$F( x ) = P ( X \leq x)\equiv P(\,\,X\in(-\infty ,x]\,\,).$$

<b>Własności dystrybuanty:</b>
1. funkcja niemalejąca, 
2. prawostronnie ciągła,
3. $\lim_{t \to -\infty} ~F( x ) = 0$, 
4. $\lim_{t \to \infty} ~F( x ) = 1$. 

<br>

    
<b>Funkcja gęstości prawdopodobieńtwa</b>

Funkcję gęstości $f$ (ang. probability density function, PDF) definiuje się tylko dla rozkładów ciągłych.

<b>Własności funkcji gęstości:</b>
1. $\int_{-\infty }^{\infty } f(x)dx = 1$
2. $F(c)\equiv P(\,\,X\in(-\infty ,c]\,\,)=\int_{-\infty }^{c}f(x)dx$
3. $P(\,\,X\in[a ,b]\,\,)=\int_{a }^{b } f(x)dx$
4. Jeśli $F(x)$ jest dystrybuantą, to $\partial F/\partial x$ jest gęstością. 
5. Wartość oczekiwaną zmiennej losowej X obliczamy ze wzoru $E(X)=\int_{-\infty }^{\infty }x f(x)dx.$

<br>
    
    
<b>Histogram</b>

Histogram jest przybliżoną reprezentacją rozkładu danych liczbowych, a dokładniej estymatorem funkcji gęstości prawdopodobieństwa. Aby skonstruować histogram, pierwszym krokiem jest podział zakresu wartości na serię przedziałów ("kubełków", ang. bins), a następnie policzenie, ile wartości przypada na każdy przedział. 
    
<br>
        
    
<b>Rozkład łączny i rozkłady brzegowe</b>   
    
Niech $F_{X,Y} (x, y) = P (X < x, Y < y)$  to rozkład łączny (ang. joint distribution) wektora losowego $(X,Y)$. Wtedy rozkłady brzegowe (ang. marginal distribution) to:
$$F_X (x) = P (X < x) = P (X < x, Y < \infty) = lim_{y\to\infty} F_{X,Y} (x, y),$$
$$F_Y (y) = P (Y < y) = P (X < \infty, Y < y) = lim_{x\to\infty} F_{X,Y} (x, y).$$
    
Jeśli $f( x , y )$ jest gęstością wektora losowego (X, Y) o znanym łącznym rokładzie prawdopodobieństwa, to rozkłady brzegowe otrzymujemy ze wzorów $$ f_{X}(x)=\int_{-\infty }^{\infty } f(x,y)\,dy,$$  $$ f_{Y}(y)=\int_{-\infty }^{\infty } f(x,y)\,dx.$$
    
<br>
        
    
<b>Niezależne zmienne losowe</b>   
    
Ogólnie, zmienne losowe X i Y są niezależne wtedy i tylko wtedy, gdy ich łączny rozkład prawdopodobieństwa spełnia warunek
    $$P ( X = x,\,   Y = y ) = P ( X = x ) \cdot P ( Y = y )$$ 
dla każdej wartości x i y.

Ciągłe zmienne losowe X i Y są niezależne wtedy i tylko wtedy, gdy  funkcja gęstości łącznego rozkładu prawdopodobieństwa wektora losowego (X, Y) jest iloczynem rozkładów brzegowych:
    $$ f_{X,Y}(x,y)=f_{X}(x)\cdot f_{Y}(y).$$

<b>Rozkład jednostajny dyskretny</b>

Rozkład jednostajny dyskretny, zwany także równomiernym, jest symetrycznym rozkładem prawdopodobieństwa, w którym skończona liczba wartości jest równie prawdopodobna. Każda z $n$ wartości ma równe prawdopodobieństwo $1/n$. Oznaczenie $U(a,b)$ oznacza, że wartości, które przyjmuje zmienna losowa pochodzą z przedziału $[a,b]$. Na przykład rzut kostką o sześciu bokach z równym prawdopobobieństwem wypadnięcia  od 1 do 6 oczek.

 Rozkład $X \sim U(a,b)$: <br>
 $$P ( X = a_1 ) = 1/n$$ 
 $$P ( X = a_2 ) = 1/n$$
 $$\vdots$$
 $$P ( X = a_n ) = 1/n$$
 
 * Wartość oczekiwana $EX = \frac{a+b}{2}$<br>
 * Wariancja $Var(X) = \frac{n^2-1}{12}$<br>
 <br>


<b>Rozkład Bernoulliego</b>

Rozkład Bernoulliego, nazwany na cześć szwajcarskiego matematyka Jacoba Bernoulliego, jest dyskretnym rozkładem prawdopodobieństwa zmiennej losowej, która przyjmuje wartość 1 z prawdopodobieństwem $p$ oraz wartość 0 z prawdopodobieństwem $q=1-p$. Na przykład rozkład prawdopodobieństwa w przypadku rzutu oszukaną (niewyważoną) monetą, dla której prawdopodobieństwa wypadnięcia orła i reszki nie są równe (gdyby były równe, to $p=q=1/2$).

 Rozkład $X \sim Bernoulli(p)$: <br>
 $$P ( X = 1 ) = p$$ 
 $$P ( X = 0 ) = 1 − p$$
 
 * Wartość oczekiwana $EX = p$ <br>
 * Wariancja $Var(X) = p(1-p)$ <br>
 <br>
 

<b>Rozkład dwumianowy</b>

Rozkład dwumianowy z parametrami $n$ i $p$ jest dyskretnym rozkładem prawdopodobieństwa liczby $k$ "sukcesów" w sekwencji $n$ niezależnych losowań. Każde losowanie ma własny wynik o wartości logicznej: "sukces" (z prawdopodobieństwem $p$) lub "porażka" (z prawdopodobieństwem $q = 1 − p$). Pojedynczy eksperyment dotyczący sukcesu/porażki (tj. $n = 1$) ma rozkład Bernoulliego. 

 Rozkład $\sum_{i=1}^{n} X_i \sim binomial(p,n)$: <br>
 $$P ( \sum_{i=1}^{n} X_i = k ) = \binom{n}{k} p^k q^{n-k},$$
 gdzie ciąg zmiennych losowych $X_1,X_2,\dots,X_n$ dla $i=1,\dots,n$ ma rozkład Bernoulliego:
 $$P ( X_i = 1 ) = p$$ 
 $$P ( X_i = 0 ) = q$$
 
 * Wartość oczekiwana $EX = np$ <br>
 * Wariancja $Var(X) = np(1-p)$ <br>
 <br>
 
 
<b>Rozkład Poissona</b>

Rozkład Poissona, nazwany na cześć francuskiego matematyka Siméona Denisa Poissona, jest dyskretnym rozkładem prawdopodobieństwa, który wyraża prawdopodobieństwo wystąpienia określonej liczby zdarzeń w określonym przedziale czasu, jeśli zdarzenia te występują ze znaną średnią  częstotliwością i w sposób niezależny od czasu, jaki upłynął od ostatniego zajścia takiego zdarzenia.

 Rozkład $X \sim Poisson(\lambda)$: <br>
 Jeśli oczekiwaną liczbą zdarzeń w pewnym przedziale jest $\lambda$, to prawdopodobieństwo, że jest dokładnie $k$ wystąpień ($k=0,1,2,\dots$) jest równe:
 $$P(X=k)={\frac {\lambda ^{k}e^{-\lambda }}{k!}}$$ 

 * Wartość oczekiwana $EX = \lambda$ <br>
 * Wariancja $Var(X) = \lambda$ <br>
 <br>

<b>(b) Bank</b> $\color{red}{\textbf{(2 p.)}}$  <br>W ciągu godziny do oddziału banku przychodzi średnio 4.2 osoby. Sprawdź prawdopodobieństwo, że:

```python
# KAWIARNIA

# **Rozkład Poissona**
# size - liczba losowanych liczb
# mu - parametr rozkładu (lambda)

# import poisson distribution
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

k = np.arange(0, 17)
lambd = 7
pmf = poisson.pmf(k, mu=lambd) # probability mass function

print("Prawdopodobieństwo wystąpienia k zdarzeń:")
for val, prob in zip(k,pmf):
    print(f"P(k={val}) = {prob.round(4)}")
    
    
suma = pmf[:11].sum()
print('\nPrawdopodobieństwo, że w kawiarni będzie mniej niż 11 klientów:')
print(f"P(0<=k<=10) = {suma.round(4)}")    

print('\nPrawdopodobieństwo, że w kawiarni będzie więcej niż 10 klientów:')
print(f"P(k>10) = {(1-suma).round(4)}")    


plt.plot(k, pmf, marker='o')
plt.xlabel('k')
plt.ylabel('Prawdopodobieństwo')
plt.title(f"KAWIARNIA: Rozkład Poisson( k, $\lambda$ ={lambd} )")
plt.show()
```

<b>Rozkład jednostajny ciągły</b>

Rozkład jednostajny ciągły, zwany także równomiernym. Rozkład jednostajny opisuje losowanie, w którym istnieje arbitralny wynik mieszczący się w pewnych granicach. Granice są określone przez parametry $a$ i $b$, które są wartościami minimalnymi i maksymalnymi przyjmowanymi przez zmienną losową.

 Gęstość $X \sim U(a,b)$: <br>
 $$ f(x;a , b)={\begin{cases}{\frac {1}{b-a}}&\mathrm {dla} \ a\leq x\leq b,
 \\[8pt]0&\mathrm {w~p.w.}\end{cases}}$$ 
 
 * Wartość oczekiwana $EX = \frac{a+b}{2}$<br>
 * Wariancja $Var(X) = \frac{(b-a)^2}{12}$<br>
 <br>


<b>Rozkład normalny (Gaussowski)</b>

Rozkład normalny to rozkład prawdopodobieństwa używany do modelowania zjawisk, które mają pewne domyślne zachowanie (średnie) i skumulowane możliwe odchylenia od tego zachowania. Na przykład, oczekuje się, że strzały wprawnego łucznika wylądują wokół dziesiątki na tarczy. Jednakże, ze względu na agregację niedoskonałości w technice łucznika, większość strzał nie trafi w dziesiątkę. Średnia tej odległości jest znana w łucznictwie jako _dokładność_, a wielkość zmienności odległości jako _precyzja_. W kontekście rozkładu normalnego dokładność i precyzję określa się odpowiednio jako _średnią_ i _odchylenie standardowe_. Biegłość łucznika można wyrazić dwoma wartościami: średnią $\mu$ i odchyleniem standardowym $\sigma$. W rozkładzie normalnym te dwie wartości oznaczają, że: istnieje ~68% prawdopodobieństwo, że strzała wyląduje w granicach jednego odchylenia standardowego od średniej celności łucznika; ~95% prawdopodobieństwa, że strzała wyląduje w granicach dwóch standardowych odchyleń od średniej celności łucznika; ~99,7% w ciągu trzech; i tak dalej, powoli wzrastając do 100%. 

 Gęstość $X \sim N(\mu,\sigma^2)$: <br>
 $$ f(x;\mu ,\sigma)={\frac {1}{\sigma {\sqrt {2\pi }}}}e^{-{\frac {1}{2}}\left({\frac {x-\mu }{\sigma }}\right)^{2}} $$ 
 
 * Wartość oczekiwana $EX = \mu$ <br>
 * Wariancja $Var(X) = \sigma^2$ <br>
 <br>
 

<b>Rozkład wykładniczy</b>

Jeżeli w jednostce czasu zachodzi średnio $\lambda $ niezależnych zdarzeń, to rozkład wykładniczy opisuje odstępy czasu pomiędzy kolejnymi zdarzeniami. 

 Gęstość $X \sim Exp(\lambda)$: <br>
 $$ f(x;\lambda )={\begin{cases}\lambda e^{-\lambda x}&x\geq 0,\\0&x<0.\end{cases}} $$
 
 * Wartość oczekiwana $EX = \frac{1}{\lambda}$<br>
 * Wariancja $Var(X) = \frac{1}{\lambda^2}$<br>
 <br>



<b>Rozkład gamma</b>

Rozkład gamma jest dwuparametrową rodziną ciągłych rozkładów prawdopodobieństwa. Rozkład wykładniczy, rozkład Erlanga i rozkład chi-kwadrat to szczególne przypadki rozkładu gamma. 

 Gęstość $X\sim \Gamma (\alpha ,\beta )$:
 $$f(x;\alpha ,\beta )={\frac {x^{\alpha -1}e^{-\beta x}\beta ^{\alpha }}{\Gamma (\alpha )}}\quad {\text{ dla }}x>0\quad \alpha ,\beta >0$$
 
 gdzie funkcja gamma $\Gamma (\alpha )=(\alpha -1)!$ dla $\alpha \in  \mathrm{N}$.
 
 * Wartość oczekiwana $EX = \frac{\alpha}{\beta}$<br>
 * Wariancja $Var(X) = \frac{\alpha}{\beta^2}$<br>
 <br>


<b>Rozkład beta</b>

Rozkład beta również korzysta z funkcji gamma. 

 Gęstość $X\sim B(\alpha ,\beta )$:
 $$f(x;\alpha ,\beta )=\frac{\Gamma (\alpha + \beta)}{\Gamma (\alpha)\Gamma (\beta)} x^{\alpha -1}(1-x)^{\beta-1}\quad {\text{ dla }}0<x<1\quad \alpha,\beta >0.$$

 
 * Wartość oczekiwana $EX = \frac{\alpha}{\alpha+\beta}$<br>
 * Wariancja $Var(X) = \frac{\alpha \beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$<br>
 <br>

```python
# **Dystrybuanta**
# N(m,s^2) 
# funkcja cdf(x)

from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

x = np.linspace(-6,6) # wartości na osi OX

#Display the cumulative distribution function (cdf):

m=0
s=1
ax.plot(x, norm.cdf(x, loc=m, scale=s), 'k-', lw=2, label='rys. 1')
m=1
s=1
ax.plot(x, norm.cdf(x, loc=m, scale=s), 'b-', lw=2, label='rys. 2')
m=2
s=1
ax.plot(x, norm.cdf(x, loc=m, scale=s), 'b--', lw=2, label='rys. 3')
m=0
s=1.5
ax.plot(x, norm.cdf(x, loc=m, scale=s), 'r-', lw=2, label='rys. 4')
m=0
s=2.5
ax.plot(x, norm.cdf(x, loc=m, scale=s), 'r--', lw=2, label='rys. 5')

ax.legend(loc='best', frameon=True)
ax.set(xlabel='wartość zm. los.', title='Dystrybuanta rozkładu normalnego $N(m, s^2)$')
plt.show()
```

```python
# **Rozkład normalny**
# size - liczba losowanych liczb
# bins - liczba przedziałów wartości histogramu
# loc - średnia
# scale - odchylenie standardowe

from scipy.stats import norm
import matplotlib.pyplot as plt
plt.style.use('default')
fig, ax = plt.subplots(1, 1)


# próba losowa N(0,1)
sample_size = 10000
data_normal = norm.rvs(size=sample_size,loc=0,scale=1)


# histogram
bins = 200 # ile przedziaów histogramu
ax.hist(data_normal, bins=bins) 
ax.set(xlabel='Sample value', ylabel='Numner of samples', 
        title=f'Histogram N(0,1), sample size={sample_size}, bins={bins}', xlim=[-8,8])

```
